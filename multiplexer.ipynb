```python
from qiskit import QuantumCircuit, QuantumRegister
import qiskit.tools.jupyter
%monospaced_output
```

```python
i = QuantumRegister(2, 'i')
s = QuantumRegister(1, 's')
g = QuantumRegister(3, 'g')
C = QuantumCircuit(i, s, g)
```

```python
C.ccx(i[1], s, g[0])
C.x(g[0])
C.ccx(i[0], g[0], g[1])
C.x(g[0])
C.x(g[1])
C.ccx(g[0], g[1], g[2])
C.x(s)
```

```python
C.draw('mpl')
```
