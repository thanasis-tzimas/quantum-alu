```python
from qiskit import QuantumCircuit, QuantumRegister
```

```python
a = QuantumRegister(1, 'a')
b = QuantumRegister(1, 'b')
eq = QuantumRegister(1, 'a=b')
lt = QuantumRegister(1, 'a<b')
gt = QuantumRegister(1, 'a>b')
C = QuantumCircuit(a, b, eq, lt, gt)
```

```python
C.cx(a, b)
C.cx(b, eq)
C.cx(a, b)
C.x(eq)
C.x(a)
C.ccx(a, b, lt)
C.x(a)
C.x(b)
C.ccx(a, b, gt)
C.x(b)
```

```python
C.draw('mpl')
```
