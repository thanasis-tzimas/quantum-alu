```python
from qiskit import QuantumCircuit, QuantumRegister
```

```python
a = QuantumRegister(1, 'a')
b = QuantumRegister(1, 'b')
c = QuantumRegister(1, 'c')
d = QuantumRegister(1, 'd=0')
m = QuantumRegister(1, 'm')
C = QuantumCircuit(a, b, c, d, m)
```

```python
C.cx(m, b)
C.ccx(a, b, c)
C.cx(a, b)
C.ccx(b, c, d)
C.cx(b, c)
C.cx(a, b)
```

```python
C.draw('mpl')
```
