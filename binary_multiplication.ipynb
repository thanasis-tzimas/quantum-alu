```python
from qiskit import QuantumCircuit, QuantumRegister
```

```python
a = QuantumRegister(2, 'A')
b = QuantumRegister(2, 'B')
o = QuantumRegister(9, 'o')
C = QuantumCircuit(a, b, o)
```

```python
C.ccx(a[0], b[0], o[0])
C.ccx(a[0], b[1], o[1])
C.ccx(a[1], b[0], o[2])
C.ccx(a[1], b[1], o[3])
```

```python
C.cx(o[1], o[4])
C.cx(o[2], o[5])
C.cx(o[3], o[7])
```

```python
C.ccx(o[4], o[5], o[6])
C.cx(o[4], o[5])
```

```python
C.ccx(o[6], o[7], o[8])
C.cx(o[6], o[7])
```

```python
C.draw('latex')
```
