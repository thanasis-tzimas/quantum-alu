```python
from qiskit import QuantumCircuit, QuantumRegister
```

```python
a = QuantumRegister(2, 'A')
b = QuantumRegister(2, 'B')
anc = QuantumRegister(9, 'anc')
C = QuantumCircuit(a, b, anc)
```

```python
C.ccx(a[0], b[0], anc[0])
C.ccx(a[0], b[1], anc[1])
C.ccx(a[1], b[0], anc[2])
C.ccx(a[1], b[1], anc[3])
```

```python
C.cx(anc[1], anc[4])
C.cx(anc[2], anc[5])
C.cx(anc[3], anc[7])
```

```python
C.ccx(anc[4], anc[5], anc[6])
C.cx(anc[4], anc[5])
```

```python
C.ccx(anc[6], anc[7], anc[8])
C.cx(anc[6], anc[7])
```

```python
C.draw('latex')
```
