```python
from qiskit import QuantumCircuit, QuantumRegister
import qiskit.tools.jupyter
```

```python
i = QuantumRegister(2, 'I')
s = QuantumRegister(1, 'S')
anc = QuantumRegister(3, 'anc')
C = QuantumCircuit(i, s, anc)
```

```python
C.ccx(i[1], s, anc[0])
C.x(anc[0])
C.ccx(i[0], anc[0], anc[1])
C.x(anc[0])
C.x(anc[1])
C.ccx(anc[0], anc[1], anc[2])
C.x(s)
```

```python
C.draw('latex')
```
